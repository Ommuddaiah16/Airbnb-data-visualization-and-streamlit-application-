```python
myclient = pymongo.MongoClient("mongodb+srv://**********h:*******@cluster0.ewofdbh.mongodb.net/")
mydb = myclient["Airbnb"]
coll = mydb["airbnb_listings"]
```

```python
data= []

for i in coll.find({},{"id":1, "listing_url":1, "name":1, 'property_type':1, 'room_type':1, 'bed_type':1, 'minimum_nights':1, 'maximum_nights':1, 'cancellation_policy':1, 'accommodates':1, 'bedrooms':1, 'beds':1, 'number_of_reviews':1, 'bathrooms':1, 'price':1, 'extra_people':1, 'guests_included':1, 'picture_url':1, 'review_scores_rating':1,'cleaning_fee':1}):
    data.append(i)
```

```python
data
df_1 = pd.DataFrame(data)
df_1.head(4)
```

```python
df_1.describe().T
```

```python
df_1['name'].fillna('unknown', inplace = True)
df_1['property_type'].fillna('Not specified', inplace = True)
df_1["beds"].fillna(0,inplace= True)
df_1["bedrooms"].fillna(0,inplace= True)
df_1["bathrooms"].fillna(0,inplace= True)
df_1["cleaning_fee"].fillna(0,inplace= True)
df_1["review_scores_rating"].fillna(0,inplace= True)
df_1.isnull().sum()
```

```python
df_1.dtypes
```

```python
# changing data types

df_1["bedrooms"]= df_1["bedrooms"].astype(int)
df_1["beds"]= df_1["beds"].astype(int)
df_1["bathrooms"]= df_1["bathrooms"].astype(str).astype(float).astype(int)
df_1["price"] = df_1["price"].replace({'\$': '', ',': ''}, regex=True)
df_1["price"] = df_1["price"].astype(float).astype(int)
df_1["extra_people"] = df_1["extra_people"].replace({'\$': '', ',': ''}, regex=True)
df_1["extra_people"]= df_1["extra_people"].astype(str).astype(float).astype(int)
df_1["guests_included"]= df_1["guests_included"].astype(str).astype(float).astype(int)
df_1["cleaning_fee"] = df_1["cleaning_fee"].replace({'\$': '', ',': ''}, regex=True)
df_1["cleaning_fee"]= df_1["cleaning_fee"].astype(str).astype(float).astype(int)
df_1["review_scores_rating"]= df_1["review_scores_rating"].astype(int)
df_1.dtypes
```

**Collecting Host Details**

```python
host= []
for i in coll.find({},{'id':1,'host_id':1, 'host_url':1, 'host_name':1, 'host_location':1,"host_response_time":1, 'host_thumbnail_url':1, 'host_picture_url':1, 'host_neighbourhood':1, 'host_response_rate':1, 'host_is_superhost':1, 'host_has_profile_pic':1, 'host_identity_verified':1, 'host_listings_count':1,'host_verifications':1}):
    host.append(i)
```

```python
df_host_1 = pd.DataFrame(host)
df_host_1.head(2)
```

```python
df_host_1.isnull().sum()
```

```python
#filling Missing data

df_host_1['host_name'].fillna('unknownn', inplace = True)
df_host_1['host_location'].fillna("Not Specified",inplace= True)
df_host_1['host_response_time'].fillna("Not Specified",inplace= True)
df_host_1['host_response_rate'].fillna("0",inplace= True)
df_host_1['host_is_superhost'].fillna("Not Specified",inplace= True)
df_host_1['host_thumbnail_url'].fillna("Not Specified",inplace= True)
df_host_1['host_picture_url'].fillna("Not Specified",inplace= True)
df_host_1['host_neighbourhood'].fillna("Not Specified",inplace= True)
df_host_1['host_listings_count'].fillna(0,inplace= True)
df_host_1['host_has_profile_pic'].fillna("Flase",inplace= True)
df_host_1['host_identity_verified'].fillna("Flase",inplace= True)

df_host_1.isna().sum()
```

```python
# Changing the "True" or "False" features

df_host_1["host_is_superhost"]= df_host_1["host_is_superhost"].map({False: "No", True: "Yes"})
df_host_1["host_has_profile_pic"]= df_host_1["host_has_profile_pic"].map({False: "No", True: "Yes"})
df_host_1["host_identity_verified"]= df_host_1["host_identity_verified"].map({False: "No", True: "Yes"})
```

**Collecting Address Details**

```python
# extracting address details

address= []
for i in coll.find({}, {'id':1, 'street':1, 'market':1, 'country':1,
                   'country_code':1, "longitude":1, "latitude":1,
                    "is_location_exact":1}):
    address.append(i)
```

```python
df_address_1= pd.DataFrame(address)
df_address_1["is_location_exact"]= df_address_1["is_location_exact"].map({False: "No", True: "Yes"})
df_address_1.head(2)
```

```python
availability= []
for i in coll.find({}, {'id':1, 'availability_30':1, 'availability_60':1, 'availability_90':1, 'availability_365':1}):
    availability.append(i)
```

```python
df_availabe_1 = pd.DataFrame(availability)
df_availabe_1.head(4)
```

**Collecting Ameities**

```python
amenities= []
for i in coll.find({},{"id":1, "amenities":1}):
    amenities.append(i)
```

```python
df_amenities =pd.DataFrame(amenities)
df_amenities
```

**Merging Data Sets**

```python
#droppping column _id

df_1.drop(columns=["_id"], inplace=True)
df_host_1.drop(columns=["_id"], inplace=True)
df_address_1.drop(columns=["_id"], inplace=True)
df_availabe_1.drop(columns=["_id"], inplace=True)
df_amenities.drop(columns=["_id"], inplace=True)
```

```python
df= pd.merge(df_1, df_host_1, on="id")
df= pd.merge(df, df_address_1, on="id")
df= pd.merge(df, df_availabe_1, on="id")
df= pd.merge(df, df_amenities, on="id")


```

```python
df.head(4)
```

```python
df.dtypes
```

```python
# Fill missing values for host_has_profile_pic, host_identity_verified, and host_is_superhost
fill_cols = ['host_has_profile_pic', 'host_identity_verified', 'host_is_superhost']
for col in fill_cols:
    df[col].fillna('not specified', inplace=True)


# Replace "N/A" with "not specified" in host_response_rate
df['host_response_rate'] = df['host_response_rate'].replace('N/A', 'unknown')
```
